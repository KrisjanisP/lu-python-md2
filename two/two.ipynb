```python
import pandas as pd
path = 'gaisaTemperatura2022.xlsx'
df = pd.read_excel(path)
print(df.head())
```

```python
averages = df.iloc[:, 1:].mean(axis=1)
df['average'] = averages
for i in range(0, 24):
    name = f"{str(i).zfill(2)}:00"
    del df[name]
print(df.head())
```

```python
# convert 'Datums' column to seaso
print(list(zip(df['Datums'], df['average'])))

winter = []
spring = []
summer = []
autumn = []

for date, temp in zip(df['Datums'], df['average']):
    if '.01.' in date or '.02.' in date or '.12.' in date:
        winter.append(temp)
    elif '.03.' in date or '.04.' in date or '.05.' in date:
        spring.append(temp)
    elif '.06.' in date or '.07.' in date or '.08.' in date:
        summer.append(temp)
    elif '.09.' in date or '.10.' in date or '.11.' in date:
        autumn.append(temp)

winter.sort()
spring.sort()
summer.sort()
autumn.sort()

print(winter)
print(spring)
print(summer)
print(autumn)
```

```python
import matplotlib.pyplot as plt

plt.boxplot([winter, spring, summer, autumn],patch_artist=True, boxprops=dict(facecolor='#2980b9'))
plt.ylabel('Gaisa temperatūra (Celsija grādi)')
plt.xticks([1,2,3,4],['ziema', 'pavasaris', 'vasara', 'rudens'])  # Replace 'Label1' with your label
plt.title('Gaisa temperatūra Rīgā četros gadalaikos')
plt.show()
```
