```python
import pandas as pd
faktiskais_path = 'vejaAtrumsFaktiskais.xlsx'
faktiskais_df = pd.read_excel(faktiskais_path)
print(faktiskais_df.head())
brazmas_path = 'vejaAtrumsBrazmas.xlsx'
brazmas_df = pd.read_excel(brazmas_path)
print(brazmas_df.head())
```

```python
faktiskais_averages = faktiskais_df.iloc[:, 1:].mean(axis=1)
brazmas_maximas = brazmas_df.iloc[:, 1:].max(axis=1)
print(faktiskais_averages)
print(brazmas_maximas)
```

```python
x, y_fakt, y_brazm = [], [], []
for i in range(len(faktiskais_averages)):
    x.append(f"{str(i).zfill(2)}.08.2023")
    y_fakt.append(faktiskais_averages[i])
    y_brazm.append(brazmas_maximas[i])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.bar(x, y_brazm, label='maks. vēja ātrums')
ax.bar(x, y_fakt, label='vid. vēja ātrums')
ax.set_xticks(x, labels=x, rotation=60)

ax.set_ylabel('vēja ātrums (m/s)')
ax.set_xlabel('mērīju datums')
ax.set_title('vidējais un maksimālais vēja ātrums 2023. g. augustā')
ax.legend()

plt.show()
```
